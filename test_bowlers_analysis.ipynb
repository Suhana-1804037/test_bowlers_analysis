```python
df = pd.read_csv("wickets.csv")

display(df.head(10))
```

# Remove BBI and BBM columns

```python
df = df.drop('BBI', axis=1)
df.drop('BBM', axis = 1, inplace = True)
display(df.head(5))
```

## Splitting Player column into Player and Country and Concatenate

```python
#splitting dataframe 
df_player = df['Player'].str.split("(", expand=True)

#concatenating df and df-player dataframe
df = pd.concat([df, df_player], axis=1)

#droping 'Player' column from the df dataframe
df = df.drop('Player', axis=1)

#renaming column o to Player and 1 to Country
df = df.rename(columns={0: 'Player',
                        1: 'Country'})

#replacing ')'(exact character ) 
df['Country'] = df['Country'].str.replace(")", "")

#creating set of new column sequence
new_col_seq = ['Player', 'Span', 'Mat', 'Inns', 'Balls', 'Runs', 'Wkts', 'Ave', 'Econ', 'SR', '5', '10', 'Country' ]

#putting the new column sequence to df
df = df[new_col_seq]

display(df.head(5))
```

# Number of players played for ICC 

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['Played_for_ICC'] = df['Country'].apply(icc_check)
display(df.head(10))
```

```python
df['Played_for_ICC'].value_counts()
```

# Number of unique countries

```python
# Removing "ICC/" and "/ICC" from the 'Country'

df['Country'] = df['Country'].str.replace("ICC/", "")
df['Country'] = df['Country'].str.replace("/ICC", "")

display(df.head(10))
```

```python
print(df['Country'].unique())
print(len(df['Country'].unique()))
```

# Number of years played by the bowlers

```python
#splitting the 'Span' column based on the "-"
df_span = df['Span'].str.split("-", expand=True)

# conccatinating the new dataframe with the main dataframe
df = pd.concat([df, df_span], axis=1)

# renaming the newly created column names
df = df.rename(columns={0: "Start_year",
                        1: "End_year"})

# removing the "Span" column
df = df.drop('Span', axis=1)

# Data type conversion
df['Start_year'] = df['Start_year'].astype('int') 
df['End_year'] = df['End_year'].astype('int')

# Counting the Numbers of years played by the playes 
df['years_played'] = df['End_year'] - df['Start_year']
df = df.drop(['Start_year', "End_year"], axis=1)

display(df.head(10))
```
